# Работа со стоп-словами

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
def remove_stopwords(text, language='russian'):
    stop_words = set(stopwords.words(language))

    # Простая токенизация без word_tokenize
    import re
    tokens = re.findall(r'\b\w+\b', text.lower())

    filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]

    return ' '.join(filtered_tokens)
```

```python
if __name__ == "__main__":

    file_path = "/content/Дневник_экспедиции_без_эмоджи.txt"

with open(file_path, "r", encoding="utf-8") as f:
    text = f.read()

    cleaned_text = remove_stopwords(text, language='russian')

    print("Исходный текст:\n", text)
    print("\Очищенный текст:\n", cleaned_text)
```

```python
with open('Дневник_экспедиции_без_стоп-слов.txt', 'w', encoding='utf-8') as f:
    f.write(cleaned_text)
```

# Лемматизация

PyMorphy3

```python
# Установка библиотек
!pip install pymorphy3 nltk

import nltk
nltk.download('punkt_tab')

import pymorphy3
```

```python
# Лемматизация
def lemmatize_text(text: str):
    morph = pymorphy3.MorphAnalyzer()
    tokens = nltk.word_tokenize(text, language='russian')

    lemmas = []
    for token in tokens:
        parsed = morph.parse(token)[0]
        lemmas.append(parsed.normal_form)

    return lemmas
```

```python
# Загрузка текста по пути
file_path = "/content/Дневник_экспедиции_без_стоп-слов.txt"

with open(file_path, "r", encoding="utf-8") as f:
    text = f.read()

print("Исходный текст:")
print(text)
```

```python
# Лемматизация
lemmas = lemmatize_text(text)

print("\nСписок лемм:")
print(lemmas)
```

```python
with open('/content/Дневник_экспедиции_lemmas_pymorphy3.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(lemmas))

print("\nЛеммы успешно сохранены в файл!")

print("\nЛемматизированная строка:")
print(" ".join(lemmas))
```

```python
# Python-библиотека Razdel — часть проекта Natasha, делит русскоязычный текст на токены и предложения.
!pip install razdel
```

```python
# загружаем модуль tokenize
from razdel import tokenize
```

```python
# указываем язык
stemmer = SnowballStemmer("russian")
```

```python
# Чтение текста из файла
with open('/content/Дневник_экспедиции_lemmas_pymorphy3.txt', 'r', encoding='utf-8') as f:
    ru_text = f.read()
```

# Присвоение грамматических тегов

Stanza

```python
!pip install stanza
```

```python
import stanza
stanza.download('ru')
nlp = stanza.Pipeline(lang='ru', processors='tokenize,pos', use_gpu=False)
```

```python
with open('/content/Дневник_экспедиции_без_стоп-слов.txt', 'r', encoding='utf-8') as f:
    text = f.read()

file_path = '/content/Дневник_экспедиции_без_стоп-слов.txt'
doc = nlp(text)

# Формируем список строк с тегами
lines = []
for sent in doc.sentences:
    for word in sent.words:
        upos = word.upos if hasattr(word, 'upos') and word.upos else '_'
        feats = word.feats if hasattr(word, 'feats') and word.feats else '_'
        # word.text — исходный токен; word.lemma — лемма (если нужна)
        lemma = word.lemma if hasattr(word, 'lemma') and word.lemma else '_'
        lines.append(f"word:\t{word.text}\tlemma:\t{lemma}\tupos:\t{upos}\tfeats:\t{feats}")
```

```python
# Склеиваем в одну строку и сохраняем
out_text = "\n".join(lines)
out_path = '/content/Дневник_экспедиции_tags_stanza.txt'

with open(out_path, 'w', encoding='utf-8') as f:
    f.write(out_text)
```

PyMorphy3

```python
!pip install pymorphy3 razdel

import pymorphy3
from razdel import tokenize
```

```python
# Инициализация морфоанализатора
morph = pymorphy3.MorphAnalyzer()
```

```python
# Чтение файла
with open('/content/Дневник_экспедиции_без_стоп-слов.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
# Токенизация
tokens = list(tokenize(text))
```

```python
# Создаем файл для результатов
with open('Дневник_экспедиции_pymorphy_анализ.txt', 'w', encoding='utf-8') as f:

    # Записываем заголовок
    f.write("ТОКЕН | НАЧАЛЬНАЯ ФОРМА | ЧАСТЬ РЕЧИ | ГРАММЕМЫ\n")
    f.write("-" * 70 + "\n")

    # Анализируем каждый токен
    for token in tokens:
        word = token.text

        # Пропускаем пробелы и пустые токены
        if not word or word.isspace():
            continue

        # Морфологический анализ
        parsed = morph.parse(word)[0]

        # Получаем часть речи
        pos = parsed.tag.POS
        pos_str = str(pos) if pos else "НЕОПР"

        # Получаем граммемы на русском
        grammemes = parsed.tag.cyr_repr

        # Записываем результат
        f.write(f"{word:<15} | {parsed.normal_form:<15} | {pos_str:<10} | {grammemes}\n")

print(f"Анализ завершен!")
print(f"Обработано токенов: {len(tokens)}")
print(f"Результаты сохранены в 'Дневник_экспедиции_pymorphy_анализ.txt'")
```
